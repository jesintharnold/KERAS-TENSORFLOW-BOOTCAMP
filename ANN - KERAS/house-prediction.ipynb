```python
#MODEL SPLITTING
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(pd_.drop("price",axis=1),pd_.price,random_state=40,test_size=0.3);
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler();

# only Train and test will be scaled , not - Y
x_train=scaler.fit_transform(x_train);
x_test=scaler.transform(x_test);
```

```python
#Importing Keras sequential and dense
from keras.models import Sequential
from keras.layers import Dense
```

```python
#3 - hidden layers with each 20 with relu activation with 1 output
model=Sequential([Dense(20,activation="relu"),Dense(20,activation="relu"),Dense(20,activation="relu"),Dense(1)]);
model.compile(optimizer="adam",loss="mse");
```

```python
model.fit(x=x_train,y=y_train,validation_data=(x_test,y_test),batch_size=50,epochs=4000);
```

```python
loss=pd.DataFrame(model.history.history);
plt.figure(figsize=(10,10))
loss.plot()
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
predict=model.predict(x_test);
```

```python
print("Mean absolute error",mean_absolute_error(y_test,predict));   # Value of by with the expected one
print("Mean squared error",mean_squared_error(y_test,predict));
```
