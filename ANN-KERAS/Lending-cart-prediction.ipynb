```python
#Analysing the Null columns seperately
# Dropping employee_length,title

data_=data_.drop(["emp_length","title"],axis=1);


sum_of_null=pd.DataFrame(data_.isnull().sum(),columns=["sum"])
sum_of_null[sum_of_null["sum"]>0]
```

```python
data_.corr()["mort_acc"]
```

```python
mort_val_=data_.groupby("total_acc").mean()["mort_acc"];


def fil_mort(total_acc,mort_val):
    if np.isnan(mort_val):
        return mort_val_[total_acc]
    else:
        return mort_val


data_["mort_acc"]=data_.apply(lambda x:fil_mort(x["total_acc"],x["mort_acc"]),axis=1)        
```

```python
data_=data_.drop(["pub_rec_bankruptcies","revol_util"],axis=1)
```

```python
# 
```

```python
data_["term"].value_counts()
data_["address"]
```

```python
data_["term"]=data_["term"].apply(lambda x:int(x[:3]));
data_["address"]=data_["address"].apply(lambda x:int(x[-5:]));
```

```python
#GRADE DROP - SUB GRADE
dummies=pd.get_dummies(data_[["sub_grade","verification_status","application_type","initial_list_status","purpose"]],drop_first=True);
data_=pd.concat([data_.drop(["sub_grade","verification_status","application_type","initial_list_status","purpose","grade"],axis=1),dummies],axis=1);
```

```python
data_["home_ownership"]=data_[ "home_ownership"].replace(["NONE","ANY"],"OTHER");
data_home_owner=pd.get_dummies(data_[["home_ownership","address","issue_d"]],drop_first=True);
data_=pd.concat([data_.drop(["home_ownership","address","issue_d"],axis=1),data_home_owner],axis=1);
data_["earliest_cr_line"]=data_["earliest_cr_line"].apply(lambda x:int(x[-4:]))
```

***Splitting and creating a model***

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(data_.drop("loan_status",axis=1),data_["loan_status"],random_state=42,test_size=0.25);
print("X-train :",x_train.shape);
print("Y-train :",y_train.shape);
print("X-test :",x_test.shape);
print("Y-test :",y_test.shape);
```

***Creating a Model using Keras Sequential***

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout
from tensorflow.keras.callbacks import EarlyStopping
```

```python
# Scaling the Input
from sklearn.preprocessing import MinMaxScaler
Scaler=MinMaxScaler();
x_train=Scaler.fit_transform(x_train);
x_test=Scaler.transform(x_test);
```

```python
model=Sequential([
    Dense(100,activation="relu"),
    Dropout(0.2),
    Dense(70,activation="relu"),
    Dropout(0.2),
    Dense(50,activation="relu"),
    Dropout(0.2),
    Dense(20,activation="relu"),
    Dropout(0.2),
    Dense(10,activation="relu"),
    Dense(1,activation="sigmoid")
]);


model.compile(optimizer="adam",loss="binary_crossentropy");
```

```python
Early_callback=EarlyStopping(patience=30,monitor="val_loss",mode="min",verbose=1);
model.fit(x=x_train,y=y_train,epochs=600,validation_data=(x_test,y_test),callbacks=[Early_callback],batch_size=200);
```

```python
__los__=pd.DataFrame(model.history.history);
__los__.plot()
```

***Evaluating the given model***

```python
from sklearn.metrics import classification_report,confusion_matrix
__pred__=model.predict(x_test);
__binary_pred__=(__pred__>0.5).astype("int32");
val=pd.DataFrame(__binary_pred__);
print(y_test.value_counts());
print(val.value_counts());
```

```python
# Classification Report and Confusion Matrix
print(classification_report(y_test,__binary_pred__))
sns.heatmap(confusion_matrix(y_test,__binary_pred__),annot=True)
```
