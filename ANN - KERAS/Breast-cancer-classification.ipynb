***Breast Cancer Detection using Keras with Early stopping Callbacks & Dropout Layers***

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('./cancer_classification.csv');
data.head()
```

***Splitting and Training/classification metrics of Breast classificaton data***

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(data.drop("benign_0__mal_1",axis=1),data["benign_0__mal_1"],test_size=0.25,random_state=42);
print("TRAIN SIZE :",x_train.shape);
print("TEST SIZE :",x_test.shape);
```

```python
#Scaling the X_train,y_train
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler();
x_train=scaler.fit_transform(x_train);
x_test=scaler.transform(x_test);
x_train
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout
from tensorflow.keras.callbacks import EarlyStopping
```

```python
model=Sequential([Dense(30,activation="relu"),Dropout(0.5),Dense(15,activation="relu"),Dropout(0.5),Dense(1,activation="sigmoid")]);
#model=Sequential([Dense(30,activation="relu"),Dense(15,activation="relu"),Dense(1,activation="sigmoid")]);
model.compile(optimizer="adam",loss="binary_crossentropy");
```

```python
Early_callback=EarlyStopping(patience=10,monitor="val_loss",mode="min",verbose=1);
model.fit(x=x_train,y=y_train,epochs=1000,validation_data=(x_test,y_test),callbacks=[Early_callback]);
#model.fit(x=x_train,y=y_train,epochs=400,validation_data=(x_test,y_test));
```

```python
loss_=pd.DataFrame(model.history.history);
loss_.describe().transpose()  
```
